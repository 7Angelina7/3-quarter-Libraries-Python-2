### Задание 1. Для чего и в каких случаях полезны различные варианты усреднения для метрик качества классификации: micro, macro, weighted?

Микро-средний показатель точности и отзыва рассчитывается на основе истинных положительных (TP), истинно отрицательных (TN), ложноположительных (FP) и ложноотрицательных (FN) индивидуальных классов модели.

Макро-средний показатель точности и запоминания рассчитывается как среднее арифметическое показателей точности и запоминания отдельных классов. Макро-средний балл F1 рассчитывается как среднее арифметическое значений F1-балла отдельных классов.

Используйте микро-усреднение , если необходимо одинаково взвесить каждый случай или прогноз.

Используйте макро-усреднение , когда нужно относиться ко всем классам одинаково, чтобы оценить общую производительность классификатора в отношении наиболее часто встречающихся меток классов.

Используйте взвешенное макро-усреднение в случае несбалансированности классов (разные случаи связаны с разными ярлыками классов). Взвешенное макросреднее рассчитывается путем взвешивания оценки каждого ярлыка класса по количеству истинных экземпляров при вычислении среднего.
Метод макро-среднего можно использовать, если вы хотите узнать, как система работает в целом по наборам данных. Вы не должны принимать какое-то конкретное решение с этим средним значением. С другой стороны, микросреднее может быть полезным показателем, когда ваш набор данных различается по размеру.

Weighted учитывает дисбаланс классов, вычисляя среднее значение двоичных показателей, в которых оценка каждого класса взвешивается по его присутствию в истинной выборке данных.

### Задание 2. В чём разница между моделями xgboost, lightgbm и catboost или какие их основные особенности?

CatBoost обладает гибкостью, позволяя задавать индексы категориальных столбцов, чтобы его можно было кодировать как 
кодирование в одно касание с использованием one_hot_max_size (используйте кодирование в одно касание для всех функций
с числом различных значений, меньшим или равным данному значению параметра).
Если вы ничего не передаете в аргументе cat_features, CatBoost будет обрабатывать все столбцы как числовые переменные.

LightGBM
Как и в CatBoost, LightGBM также может обрабатывать категориальные функции, вводя имена функций. Он не конвертируется 
в одноразовое кодирование и намного быстрее, чем одноразовое кодирование. LGBM использует специальный алгоритм,
чтобы найти значение разделения категориальных признаков.
Перед построением набора данных для LGBM вы должны преобразовать свои категориальные функции в тип int.
Он не принимает строковые значения, даже если вы передаете его через параметр categoryorical_feature.

XGBoost
В отличие от CatBoost или LGBM, XGBoost не может обрабатывать категориальные функции сам по себе,
он принимает только числовые значения, подобные случайному лесу. Поэтому перед подачей категориальных данных
в XGBoost необходимо выполнить различные кодировки, такие как кодирование меток, среднее кодирование или однократное кодирование.
